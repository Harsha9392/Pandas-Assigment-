```python
# use groupby() to find total sales by region
total_sales=data.groupby('Region')['TotalAmount'].sum()
print(total_sales)
```

```python
#display all orders placed on a weekend
data['OrderDate'] = pd.to_datetime(data['OrderDate'], errors='coerce')
weekend_orders = data[data['OrderDate'].dt.dayofweek >= 5]
print(weekend_orders)
```

```python
# Create a pivot table of TotalAmount by Region and Category.
pivot_table=data.pivot_table(values='TotalAmount',index='Region',columns='Category',aggfunc='sum',margins=True)
print(pivot_table)
```

```python
#Unpivot (melt) the pivot table from Task
pivot_reset=pivot_table.reset_index()
melted_table=pivot_reset.melt(id_vars='Region',var_name='Category',value_name='TotalAmount')
print(melted_table)
```

```python
# Use .agg() to get min, max, and mean quantity for each category.
data['Quantity'] = pd.to_numeric(data['Quantity'], errors='coerce')

data_agg_result = data.groupby('Category')['Quantity'].agg(['min','max','mean'])
print(data_agg_result)
```

```python
# Reload original data to ensure a clean slate for this merge and subsequent comparison
original_data_for_merge = pd.read_csv("/content/east_region_data.csv")
data = original_data_for_merge.copy()
customer_lookup_table = original_data_for_merge[['CustomerID', 'CustomerName']].drop_duplicates(subset=['CustomerID'])
data = pd.merge(data, customer_lookup_table.set_index('CustomerID'), on='CustomerID', how='left', suffixes=('_Original', '_Lookup'))

print(data.head())
```
